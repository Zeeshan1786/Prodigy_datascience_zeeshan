# Task 1 - Create a bar chart or histogram to visualize the distribution of a variable
# Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset directly from current Jupyter folder
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_574524.csv", skiprows=4)

# Check first 5 rows
df.head()
```

## Step 1 – Basic Info about Dataset

```python
df.info()
df.describe()
```

## Step 2 – Select Few Countries for Visualization

```python
# Your existing data prep code
countries = ["India", "China", "United States", "Brazil", "Pakistan"]
df_subset = df[df["Country Name"].isin(countries)]
df_subset = df_subset.melt(id_vars=["Country Name"], var_name="Year", value_name="Population")
df_subset["Year"] = pd.to_numeric(df_subset["Year"], errors="coerce")
df_subset = df_subset.dropna()

# Now add the plot
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

plt.figure(figsize=(36,10))
sns.barplot(data=df_subset, x="Year", y="Population", hue="Country Name")
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population of Selected Countries Over Years")
plt.xticks(rotation=45)
plt.legend(title="Country")
plt.tight_layout()
plt.show()
```
